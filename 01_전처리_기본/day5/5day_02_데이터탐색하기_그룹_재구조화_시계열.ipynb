# 2. 재구조화

```python
# cut() : 동일한 길이(값의 범위)로 3구간으로 나눈다..
# 수학점수를 기준으로 제시된 구간의 값으로 균등하게 
#  3구간으로 나누어 줍니다.
df1 = pd.cut(df["math"], 3)
df1
```

```python
# 구간별 소속을 알려줍니다. index 번호로..
df1 = pd.cut(df["math"], 3, labels=False)
df1
```

```python
df2 = df["math"].groupby(df1)
```

```python
df2.head(1)
```

```python
# df2에 대한 기초 통계..: agg() 함수 사용..
df3 = df2.agg(["count", "mean", "std", "min", "max"])
df3
```

```python
# qcut()

df
```

```python
# 수학점수의 갯수를 기준으로 제시된 구간의 값으로 
# 유사한 값들끼리(정확하게 같지는 않습니다.) 3구간으로 나누어 줍니다.
# 구간이 커지면, 해당 구간에 없는 값이 생길 수도 있습니다.
# 해당 구간에 존재하는 값이 없을 경우에는 해당 구간은 조회되지 않습니다.
pd.qcut(df["math"], 5)
```

```python
# labels=False : 범주의 번호 표시하기..
df4 = pd.qcut(df["math"], 3, labels=False)
```

```python
df4.index

for idx in df4.index :
    print(idx)
```

```python
import numpy as np

df5 = pd.qcut(df["math"], 3, labels=np.arange(3, 0, -1))
```

```python
# 첫번째값 : 시작값
# 두번째값 : 끝값 - 1
# 세번째값 : 증가값 or 감소값
for i in range(0, 3, 1) :
    print(i)

print("-------")

# 증가값을 감소 시킬때는 두번째값은 + 1
for j in range(3, 0, -1) :
    print(j)
```

```python
# 첫번째값 : 시작값
# 두번째값 : 끝값 - 1
# 세번째값 : 증가값 or 감소값
for i in np.arange(0, 3, 1) :
    print(i)
    
print("-------")

# 증가값을 감소 시킬때는 두번째값은 + 1
for j in np.arange(3, 0, -1) :
    print(j)
```

```python
df5
```

```python
df6 = df["math"].groupby(df5)
```

```python
df6.head(1)
```

```python
# 기초통계 통계 조회..
# count, mean, std, min, max

df7 = df6.agg(["count", "mean", "std", "min", "max"])
```

```python
df7
```
