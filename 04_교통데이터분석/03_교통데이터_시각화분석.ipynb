# 데이터 시각화 분석

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc
import platform
plt.rc("font" , family = "Malgun Gothic")
```

```python
# 1. 기준월 및 기준일자 별 버스 이용량 분석 비교....
# 사용할 분석요소 : 기준월, 기준일, 승객구분
# 사용할 함수 : 카운트 ....(빈도)
df_tot.head(1)
```

```python
# 피봇 테이블 만들기
df_pivot = df_tot.pivot_table(values = "승객구분",
                             index  = "기준월",
                             columns= "기준일",
                             aggfunc= "count")
df_pivot
```

```python
# 결측치 처리하기
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
# 히트맵 그리기
plt.figure(figsize=(16,10))
sns.heatmap(df_pivot, annot=True, fmt=".0f", cmap="rocket_r")
plt.title("기준월 및 기준일자별 버스 이용량 분석")
plt.show()
```

```python
# 2. 기준일별 시간대별 버스 이용량 분석
```

```python
df_pivot = df_tot.pivot_table(values = "승객구분",
                             index  = "기준일",
                             columns= "기준시간",
                             aggfunc= "count")

df_pivot
```

```python
# 결측치 처리하기
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
# 히트맵 그리기
plt.figure(figsize=(16,10))
sns.heatmap(df_pivot, annot=True, fmt=".0f", cmap="rocket_r")
plt.title("기준일별 시간대별 버스 이용량 분석")
plt.show()
```

```python
#3. 기준시간 및 분별 버스 이용량 분석
df_pivot = df_tot.pivot_table(values = "승객구분",
                             index  = "기준시간",
                             columns= "기준시간(분)",
                             aggfunc= "count")

df_pivot
```

```python
# 결측치 처리하기
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
# 히트맵 그리기
plt.figure(figsize=(16,10))
sns.heatmap(df_pivot, annot=True, fmt=".0f", cmap="rocket_r")
plt.title("기준시간 및 분별 버스 이용량 분석")
plt.show()
```

```python
# 4. 기준일별 시간대별 버스내체류시간(분) 분석
df_pivot = df_tot.pivot_table(values = "버스내체류시간(분)",
                                      index = "기준일",
                                      columns = "기준시간",
                                      aggfunc = "mean")
df_pivot
```

```python
# 결측치 처리하기
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
# 히트맵 그리기
plt.figure(figsize=(16,10))
sns.heatmap(df_pivot, annot=True, fmt=".0f", cmap="rocket_r")
plt.title("기준일별 시간대별 버스내체류시간(분) 분석")
plt.show()
```

```python
# 5. 시간대별 승객구분별 버스내 체류시간(분) 분석
```

```python
df_temp = pd.DataFrame()
df_temp["기준시간"] = df_tot["기준시간"]
df_temp["승객구분"] = df_tot["승객구분"]
df_temp["버스내체류시간(분)"] = df_tot["버스내체류시간(분)"]

df_temp
```

```python
df_temp.drop_duplicates(["승객구분"])
```

```python
# 5. 시간대별 승객구분별 버스내 체류시간(분) 분석
df_temp2 = (df_temp.groupby(["기준시간", "승객구분"], as_index=False).sum()).\
    sort_values(by=["버스내체류시간(분)"],ascending=False)
df_temp2
```

```python
df_temp2.head(10).transpose()
```

```python
fig = plt.figure(figsize = (25,10))
sns.barplot(x="기준시간", y="버스내체류시간(분)", hue="승객구분", data=df_temp2)
plt.title("시간대별 승객구분별 버스내 체류시간(분) 분석")
plt.show()
```

```python
# 시간 및 승객 구분별 빈도분석
plt.figure(figsize = (12,4))
g = sns.histplot(data=df_temp2,
                x="기준시간",
                bins =30,
                kde = True ,
                hue = "승객구분",
                multiple ="stack",
                stat = "density",
                shrink = 0.6)
plt.title("시간 및 승객구분별 빈도분석")
plt.xlabel("기준시간")
plt.ylabel("rate")
plt.show()
```

```python
# 승하차정류장별 버스내 체류시간(분) 상위 30건 분석 비교
```

```python
df_tot
```

```python
df_temp = pd.DataFrame()

df_temp["승하차정류장"] = df_tot["승차정류장"] + "-->" + \
                                    df_tot["하차정류장"] 
df_temp["승객구분"]     = df_tot["승객구분"]
df_temp["버스내체류시간"] = df_tot["버스내체류시간(분)"] 
df_temp
```

```python
# 승하차정류장(승차정류장 및 하차정류장)별 버스내체류시간 상위 30건 조회
# 승하차정류장 및 승객구분별로 버스내체류시간의 sum을 구하기..
df_temp_sel =(df_temp.groupby(["승하차정류장","승객구분"],
                             as_index = False).sum()).sort_values(by=["버스내체류시간"],
                                                                 ascending = False).head(30)
df_temp_sel
```

```python
# 그래프 사이즈 지정
plt.figure(figsize = (12,4))
# 데이터 x축 y축 넣기
plt.plot(df_temp_sel["승하차정류장"], df_temp_sel["버스내체류시간"])
# 그래프 제목
plt.title("승하차정류장별 버스내 체루시간(분) 상위 30건 분석")
# x축 제목
plt.xlabel("승하차정류장")
# y축 제목
plt.ylabel("버스내체류시간")
plt.xticks(rotation = 90)
# x축 표시값을 90도 수직으로 표현
plt.grid(True)
#그래프로 보여줘
plt.show()
```
